```python
import matplotlib.pyplot as plt
import numpy as np

def read_xvg(path):
    x = []
    y = []

    with open(path) as f:
        for line in f:
            cols = line.split()

            if len(cols) == 2:
                try:
                    x.append(float(cols[0]))
                    y.append(float(cols[1]))
                except:
                    continue
    
    return (x, y)
```

# Lysozyme tutorial figure book

Please open this only after you have completed the tutorial. The figures will not load if you have not completed the tutorial. The tutorial can be found [here](http://www.mdtutorials.com/gmx/lysozyme/index.html)

After downloading the 1AKI file, converting it to the appropriate topology, solvating and adding ions, we perform an energy minimization. This is accomplished in gromacs although you require an `mdp` file to tell gromacs what to do to perform the minimization. This is provided with the tutorial.

```python
x, y = read_xvg("potential.xvg")

fig, ax = plt.subplots(1, 1, figsize = (5, 5))
ax.plot(x,y, 'k-', lw = 2, label = 'Potential')
ax.set(ylabel = "Potential energy " + r"$( \frac{kJ}{mol} )$", xlabel = 'Energy Minimization step', title = "Energy Minimization \n using steepest descent")
```

We then perform an NVT equilibration to plateau the system temperature to stabilize the configuration and energy of the system

```python
x, y = read_xvg("temperature.xvg")

fig, ax = plt.subplots(1, 1, figsize = (5, 5))
ax.plot(x,y, 'k-', lw = 2, label = 'Potential')
ax.set(ylabel = "Temperature (K)", xlabel = 'Time (ps)', title = "Temperature of box\nduring NVT equilibration")
```

We then perform an NPT equilibration to ensure the conformation of the protein at the desired temperature and pressure is stable.

```python
sz = 5
fig, axs = plt.subplots(1, 2, figsize = (sz*2, sz))

x, y = read_xvg("pressure.xvg")
ax = axs[0]
ax.plot(x,y, 'k-', lw = 2, label = 'Potential')
ax.set(ylabel = "Pressure (bar)", xlabel = 'Time (ps)', title = "Pressure of box\nduring NPT equilibration")

x, y = read_xvg("density.xvg")
ax = axs[1]
ax.plot(x,y, 'k-', lw = 2, label = 'Potential')
ax.set(ylabel = "Density " + r"$(\frac{kg}{m^3})$", xlabel = 'Time (ps)', title = "Density of box\nduring NPT equilibration")
fig.tight_layout()
```
